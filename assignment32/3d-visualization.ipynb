```python
import random
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
N = 1000
std = 3
std_weight_melon = 100 
std_weight_baloon = 3

Melons_width = np.random.normal(10, std, N)
Melons_length = np.random.normal(16, std, N)
Melons_weight = np.random.normal(1000, std_weight_melon, N)

Balloons_width = np.random.normal(16, std, N)
Balloons_length = np.random.normal(18, std, N)
Balloons_weight = np.random.normal(5, std_weight_baloon, N)

width = np.concatenate((Melons_width, Balloons_width))
length = np.concatenate((Melons_length, Balloons_length))
weight = np.concatenate((Melons_weight, Balloons_weight))

X = np.array([width, length, weight]).T

Melons_label = np.zeros(N, dtype='int')
Balloons_label = np.zeros(N, dtype='int')

Y = np.concatenate((Melons_label, Balloons_label))
```

```python
ax = plt.axes(projection='3d')
ax.scatter(X[0:N, 0], X[0:N, 1], X[0:N, 2], c='green', label='Melon')
ax.scatter(X[N: , 0], X[N: , 1], X[N: , 2], marker= '^', c='red', label='Baloon')

ax.set_xlabel('Width')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')

ax.set_title('3D-figure')

ax.legend(loc="best")


ax.elev = 20 

plt.show()
```
