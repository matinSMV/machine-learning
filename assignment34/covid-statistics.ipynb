```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('datasets\COVID-19-time-series-clean-complete.csv')
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data["DateYearMonth"] = data["Date"].map(lambda date: 100*date.year + date.month)
data.head()
```

```python
#working on april 2020
date_april = data.query("DateYearMonth == 202004")
date_april.head()
```

```python
# sum up april case datas
april_sum = date_april.groupby(["Country/Region"])["New cases"].sum().reset_index()
april_sum.head(15)
```

```python
# show 8 country/region s
newcase_april = april_sum.sort_values("New cases")
newcase_april.tail(8)
```

```python
# mortality rate of Iran
```

```python
iran_data = data.loc[(data["Country/Region"] == "Iran")]
iran_data
```

```python
newDeaths = iran_data.groupby(["DateYearMonth"])["New deaths"].sum().reset_index()
newDeaths
```

```python
newCases = iran_data.groupby(["DateYearMonth"])["New cases"].sum().reset_index()
newCases
```

```python
plt.figure(figsize=(18,6))
plt.plot(newCases['DateYearMonth'].to_numpy(dtype=str), newCases['New cases'], marker='o', label='New Cases')
plt.plot(newDeaths['DateYearMonth'].to_numpy(dtype=str), newDeaths['New deaths'], marker='o', label='New Deaths')
plt.fill_between(newCases['DateYearMonth'].to_numpy(dtype=str), newCases['New cases'], alpha=0.5)
plt.fill_between(newDeaths['DateYearMonth'].to_numpy(dtype=str), newDeaths['New deaths'], color='orange')
plt.grid()
plt.legend(loc='best')
plt.title('New Cases vs New Deaths in IRAN')
```
